# Step 6: Join (Merge) the output of the Step 4 and Step 5. There will be no common column to join as such the join will lead to cartesian product. (Note: There will be 18 rows after the join). This step is required to create Dummy Transaction. 

```python
#Converting Date column to datetime type
Tran['Date'] = pd.to_datetime(Tran['Date'], format='%d-%b-%y')
```

```python
Tran
```

```python
Tran.info()
```

```python
#Extracting unique YYYY-MM from daate column whos datatype is datatime and storing it in distinct_yyyymm (variable)
distinct_yyyymm = Tran['Date'].dt.strftime('%Y-%m').unique()
#Extracting unique Account No and storing it in distinct_accno 
distinct_accno = Tran['Account No'].unique()
#Merging distinct_accno and distinct_yyyymm without any comman fild that why using itertools.product and saving in new name (cart_prod) with two columns name as Acc_No and Txn_Date
cart_prod = pd.DataFrame(list(itertools.product(distinct_accno, distinct_yyyymm)), columns=['Acc_No', 'Txn_Date'])
```

```python
#Output of cart_prod table
cart_prod
```

# Step 7: Do the following transformations on the Dummy Transactions created in above step:
a) The YYYY-MM dates be converted to 1st Day of the next month. For example, if the date is 2011-
04 then it should be changed to 2011-05-01 (i.e. 01-May-2011)<br>
b) Add Txn_Amt field and populate it with 0. Note: You should give appropriate names to the 
fields

```python
# convert the values in the Txn_Date column to pandas datetime objects
cart_prod["Txn_Date"] = pd.to_datetime(cart_prod["Txn_Date"])

# add an offset of 1 month to each date, and change the day to the 1st day of the month
cart_prod["Txn_Date"] = cart_prod["Txn_Date"] + pd.offsets.MonthBegin(1)

# convert the dates back to strings in the format YYYY-MM-DD
cart_prod["Txn_Date"] = cart_prod["Txn_Date"].dt.strftime("%Y-%m-%d")
```

```python
cart_prod
```

```python
#Adding Txn_Amt field and populate it with 0.
cart_prod['Txn_Amt'] = 0
```

```python
cart_prod
```

# Step 10: Aggregate the Txn_Amt values, group by Account No & Txn Date level

```python
Grp_Tran_Cart_OpeningBal = Tran_Cart_OpeningBal.groupby(['Acc_No', 'Txn_Date']).agg(Sum_Txn_Amt=('Txn_Amt', 'sum')).reset_index()
```

```python
Grp_Tran_Cart_OpeningBal
```

```python
Srt_Grp_Tran_Cart_OpeningBal = Grp_Tran_Cart_OpeningBal.sort_values(by=['Acc_No', 'Txn_Date'], ascending=[True, True])
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

# Step 11: Use the output of step 10 (10A). Create a Running Balance (Cumulative Sum) column by Account Level. For the first transaction of a given account, the value in the Running Balance field will be equal to the Txn Amount. For subsequent rows, the Running Balance = Previous Row Running Balance + Txn Amount value.

```python
Srt_Grp_Tran_Cart_OpeningBal['Running_Bal'] = Srt_Grp_Tran_Cart_OpeningBal.groupby('Acc_No')['Sum_Txn_Amt'].cumsum()
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

# Step 12: To the output of Step 11, add two columns  Prev_Txn_Date and Prev_Running_Bal

```python
Srt_Grp_Tran_Cart_OpeningBal['Prev_Txn_Date'] = Srt_Grp_Tran_Cart_OpeningBal.groupby('Acc_No')['Txn_Date'].shift(1)
Srt_Grp_Tran_Cart_OpeningBal['Prev_Running_Bal'] = Srt_Grp_Tran_Cart_OpeningBal.groupby('Acc_No')['Running_Bal'].shift(1)
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

# Step 13: Calculate Date Difference between Txn_Date and Prev_Txn_Date and store in some column (say No_Of_Days). The number of days is the period for which the previous balance was maintained in the account at each End of Day.

```python
# Convert the Txn_Date and Prev_Txn_Date columns to datetime format
Srt_Grp_Tran_Cart_OpeningBal['Txn_Date'] = pd.to_datetime(Srt_Grp_Tran_Cart_OpeningBal['Txn_Date'])
Srt_Grp_Tran_Cart_OpeningBal['Prev_Txn_Date'] = pd.to_datetime(Srt_Grp_Tran_Cart_OpeningBal['Prev_Txn_Date'])

# Calculate the time difference between Txn_Date and Prev_Txn_Date in days
Srt_Grp_Tran_Cart_OpeningBal['No_Of_Days'] = Srt_Grp_Tran_Cart_OpeningBal['Txn_Date'].diff().dt.days
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

# Step 14: Add a column Mth_ID having value in YYYYMM format. The value for this column should be computed from Prev_Txn_Date.

```python
Srt_Grp_Tran_Cart_OpeningBal['Mth_ID'] = Srt_Grp_Tran_Cart_OpeningBal['Prev_Txn_Date'].dt.strftime('%Y%m')
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

# Step 15: Calculate Month wise Average Monthly Balance by aggregating the data at Account No, YYYYMM level. While aggregating filter out the rows where Mth_ID is null. AMB = sum(Prev_Running_Bal * No_Of_Days) / sum(No_Of_Days)

```python
Srt_Grp_Tran_Cart_OpeningBal = Srt_Grp_Tran_Cart_OpeningBal.dropna(subset=['Mth_ID'])
```

```python
Srt_Grp_Tran_Cart_OpeningBal
```

```python
mthly_bal = Srt_Grp_Tran_Cart_OpeningBal.groupby(['Acc_No', 'Mth_ID']).apply(lambda x: (x['Prev_Running_Bal'] * x['No_Of_Days']).sum() / x['No_Of_Days'].sum()).reset_index(name='Avg_Mthly_Bal')
```

```python
print(mthly_bal)
```

```python
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal= pd.DataFrame(mthly_bal)
```

```python
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal
```

```python
#Add the count to the days in that month
# Extract year and month from Mth_ID column
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Year'] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Mth_ID'].str[:4]
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Month'] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Mth_ID'].str[4:]

# Convert Year and Month columns to integers
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal[['Year', 'Month']] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal[['Year', 'Month']].astype(int)

# Calculate number of days in each month based on Year and Month columns
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Cnt_Days_in_Mth'] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal.apply(lambda row: calendar.monthrange(row['Year'], row['Month'])[1] if not pd.isnull(row['Mth_ID']) else None, axis=1)

# Drop Year and Month columns
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal.drop(columns=['Year', 'Month'], inplace=True)
```

```python
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal
```

```python
#Calculating Sum_Product
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Sum_Product'] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Avg_Mthly_Bal'] * Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Cnt_Days_in_Mth']
```

```python
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal
```

```python
#Making Avg_Mthly_Bal 2 decimal
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Avg_Mthly_Bal'] = Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal['Avg_Mthly_Bal'].round(2)
```

```python
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal
```

```python
#Arranging the columns
Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal=Final_Avg_Mthly_Bal_Srt_Grp_Tran_Cart_OpeningBal[["Acc_No","Mth_ID","Sum_Product","Cnt_Days_in_Mth","Avg_Mthly_Bal"]]
```
